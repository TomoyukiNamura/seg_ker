# 　

## CSVファイル読み込み

```python
# Pandasを「pd」という略称名で読み込み
import pandas as pd
```

```python
# CSVファイルを読み込み
df = pd.read_csv('california_housing_train.csv')
df.head()
```

#  

## ヒストグラム

```python
# matplotlib.pyplotを「plt」という略称名で読み込み
%matplotlib inline
import matplotlib.pyplot as plt
```

```python
# median_house_valueのヒストグラムを描画
plt.hist(df["median_house_value"])
```

```python
# 引数でビンの数を指定可能
plt.hist(df["median_house_value"], bins=50)
```

#  

## 散布図

```python
# median_incomeとmedian_house_valueの散布図を描画
plt.scatter(df["median_income"], df["median_house_value"])
```

#  

## 散布図行列

```python
# seabornを「sns」という略称名で読み込み
import seaborn as sns
```

```python
# dfの散布図行列を描画
sns.pairplot(df)
```

#  

## 箱ひげ図

```python
# median_house_valueの箱ひげ図を描画
plt.boxplot(df["median_house_value"])
```
